```python
filtered_data_boston = []
for piece in paa_boston['data']['searchArts']['items']:
    piece_data = {
        "id": piece["id"],
        "titles": piece["titles"],
        "geo": piece["geo"],
        "address": piece["address"],
    }
    filtered_data_boston.append(piece_data)

#print(filtered_data)

#df_filter = pd.DataFrame(filtered_data)
#print(d)
```

```python
df_normal_boston= pd.json_normalize(filtered_data_boston)
df_normal_boston.rename(columns={'address.city': 'city'}, inplace=True)
df_normal_boston.head(5)
```

```python
def complete(dfs):
    art_amount=[]
    city_name=[]
    for df in dfs:
        l=len(df)
        art_amount.append(l)
        city=df['city'].iloc[0]
        city_name.append(city)
    return {'city': city_name,
     'art_amount': art_amount
    }
    #return art_amount

d = complete([df_normal_boston, df_normal_dallas, df_normal_sanfran, df_normal_nyc, df_normal_cbus, df_normal_indy])
```
